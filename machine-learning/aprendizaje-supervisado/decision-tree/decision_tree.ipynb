# Clasificación mediante un árbol de decisión

```
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

### Crear variables iniciales

Aqui para usar el csv, hay que cargarlo

```
df = pd.read_csv('bill_authentication.csv')
x = df.drop('Class',axis=1)
y = df['Class']
```

### Crear y entrenar modelo

```
model = DecisionTreeClassifier(random_state=0, max_depth=4)
model.fit(x,y)
```

### Realizar predicciones

```
model.predict([[-3,2,0,-2]])
```

### Mostrar árbol de decisión

```
import matplotlib.pyplot as plt
from sklearn import tree
```

```
f, ax = plt.subplots(1, 1, figsize = (15, 10))

tree.plot_tree(model,
           feature_names = x.columns, 
           ax=ax,
           filled = True);
```
